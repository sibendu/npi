```python
import pandas as pd

filename= 'data/out/out/npidataaa.csv'
output= 'data/out/out/out.csv'
df = pd.read_csv(filename)
df[:2]
```

```python
# Number of columns to drop
n = 275

df2 = df.iloc[:, :-n]
df2.columns = [c.lower().replace(" ", "_").replace(".", "").replace("(", "").replace(")", "") for c in df2.columns]
df2[:2]
```

```python
df2.to_csv(output)
```

```python
NEW_LINE = '\n'
dml_statement = 'CREATE TABLE NPI_DATA( ' + NEW_LINE
for c in df2.columns:
    dml_statement = dml_statement + c + ' varchar(100),' + NEW_LINE

dml_statement = dml_statement[ : len(dml_statement)-2] + NEW_LINE
dml_statement = dml_statement + ');'
print(dml_statement)
```
